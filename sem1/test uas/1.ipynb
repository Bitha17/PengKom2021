```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("WorldCups.csv")
```

```python
# Menampilkan pergerakan banyaknya penonton dari tahun ke tahun untuk data saat banyaknya peserta lebih dari 20 tim.
df.loc[df["QualifiedTeams"] > 20].plot(kind="line", x="Year", y="Attendance", color = "violet", title = "Pergerakan jumlah penonton saat peserta > 20 tim")
```

```python
# Menampilkan komposisi dalam persentase mengenai berapa kali suatu negara menjadi pemenang.
df["Winner"].value_counts().plot(kind="pie", autopct="%1.1f%%", title = "Persentase kemenangan tiap negara")
```

```python
# Membandingkan berapa kali setiap suborganisasi FIFA menjadi penyelenggara.
df["Organization"].value_counts().plot(kind="bar", title = "Frekuensi suborganisasi FIFA menjadi penyelenggara")
```

```python
# Menghitung nilai korelasi jumlah penonton dan jumlah goal
print(df["Attendance"].corr(df["GoalsScored"]))

df.plot.scatter(x="Attendance", y="GoalsScored", title = "Korelasi jumlah penonton dan jumlah goal")
```
