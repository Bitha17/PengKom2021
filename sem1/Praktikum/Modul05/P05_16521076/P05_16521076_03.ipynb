```python
# NIM/Nama  : 16521076/Tabitha Permalla
# Tanggal   : 01/12/2021
# Deskripsi : Program menampilkan (1) Histogram distribusi banyaknya review rumah yang disewakan di wilayah West Region
#             (2) Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di atas 300
#             (3) Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang
#             (4) Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari rumah
#             dengan banyak hari ketersediaan rumah dalam satu tahun di atas 100
#             (5) Stacked bar chart dari rumah yang disewakan oleh Richards dengan neighbourhood sebagai x
#             dan jenis penyediaan ruang sebagai stacked y
```

```python
# Import Library
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("P05-03.csv")
```

```python
# Histogram distribusi banyaknya review rumah yang disewakan 
# di wilayah (neighbourhood group) West Region dengan bins 20
df1 = df.loc[df["neighbourhood_group"] == "West Region"]
df1[["number_of_reviews"]].plot(kind="hist", bins = [0,20,40,60,80,100,120,140,160], rwidth = 0.8, title = "Review Rumah di West Region", color = "cyan")
```

```python
# Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di atas 300
df2 = df.loc[df["price"] > 300]
df2["neighbourhood"].value_counts().plot(kind="pie", title = "Banyak rumah per daerah dengan harga sewa per malam di atas 300")
```

```python
# Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang
df3 = df.loc[df["neighbourhood"] == "Kallang"]
df3.plot.scatter(x = "longitude", y = "latitude", title = "Distribusi Rumah di Kallang", color = "pink")
```

```python
# Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari rumah
# dengan banyak hari ketersediaan rumah dalam satu tahun di atas 100
df4 = df.loc[df["availability_365"] > 100]
df4.plot(kind="scatter",x="longitude", y="latitude",sizes=df4["number_of_reviews"], color="violet", title = "Persebaran rumah dengan ketersediaan dalam setahun > 100")
```

```python
# Stacked bar chart dari rumah yang disewakan oleh Richards dengan neighbourhood sebagai x
# dan jenis penyediaan ruang sebagai stacked y
df5 = df.loc[df["host_name"] == "Richard"]
df5.groupby(["neighbourhood","room_type"])["neighbourhood"].size().unstack().plot(kind="bar",stacked=True, title = "Rumah yang disewakan Richards")
```
