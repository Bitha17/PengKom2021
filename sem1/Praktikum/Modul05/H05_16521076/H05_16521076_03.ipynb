```python
# Import Library
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
data = "./H05.csv"
df = pd.read_csv(data)
```

```python
# Histogram distribusi nilai matematika
df[["nilai_mat"]].plot(kind="hist")
```

```python
# Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas
df["fakultas"].value_counts().plot(kind="barh", title = "Banyaknya Mahasiswa per Fakultas")
```

```python
# Diagram pie banyaknya mahasiswa masing-masing fakultas
df["fakultas"].value_counts().plot(kind="pie")
```

```python
# Menjawab soal no.4 :Berdasarkan diagram pie dan batang horizontal, manakah fakultas dengan mahasiswa terbanyak? 
#                     Manakah diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak dan mengapa?
# JAWAB : Fakultas dengan mahasiswa terbanyak adalah fakultas FTI. 
#         Diagram batang horizontal lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak.
#         Hal ini dikarenakan pada diagram pie, perbedaan jumlah antara mahasiswa FTI dan mahasiswa STEI 
#         tidak terlihat dengan jelas (karena perbedaan sudut yang kecil).
```

```python
# Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y.
df.plot.scatter(x = "nilai_kim", y = "nilai_fis")
```
